## Exercise 4: Step Activation Function 

### Objective: Convert linear output into a binary decision. 
### • Implement a simple step activation function 
### • Apply it to the linear output

```python
def stepfunc(val):
    if val > 0.5:
        return 1
    else:
        return 0
linear_output = 0.8
output = stepfunc(linear_output)
print("After Activation Function : ",output)
```

## Exercise 7 : Post Lab Reflection Questions

#### 1. What role do weights play in the dot product?

#### Weights determine the importance of each input feature in the dot product.
#### During the dot product, each input value is multiplied by its corresponding weight, so higher weight means more influence on the final output.

#### 2. How does bias affect the decision boundary?

#### Bias allows the decision boundary to shift left or right (up or down) instead of being forced to pass through the origin.
#### It provides flexibility so the model can fit the data better.

#### 3. Why is dot product fundamental to neural networks?

#### The dot product is used to calculate the weighted sum of inputs, which is the first step in every neuron.
#### Without dot product, neurons cannot combine inputs and weights to produce meaningful outputs.

#### 4. How does batch computation help in training deep learning models?

#### Batch computation processes multiple inputs at the same time, which:
#### Speeds up training
#### Uses memory efficiently
#### Stabilizes learning
#### It also allows better use of GPU/CPU parallel processing.
